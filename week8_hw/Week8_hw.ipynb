```python
def generationtester(p, population):
    i = 1
    list = []
    s = p
    while True:
        s = np.random.binomial(2*population, s)/(2*population)
        if s == 0 or s == 1:
            return(i)
            break
        else:
            i = i+1
```

```python
generationtester(0.5, 100)
```

```python
def simulation(p, population, n):
    g = []
    timer = 1
    while timer < n:
        g.append(generationtester(p, population))
        timer += 1
    return g
```

```python
t = simulation(0.5, 100, 1000)
```

```python
plt.hist(t, color = "red", bins = 60)
plt.show()
```

```python
N = []
n = 100
recorder = []
while n <= 10000000:
    recorder.append(generationtester(0.5, n))
    N.append(n)
    print(n)
    if n <= 1000000:
        n += 10000
    if n > 1000000:
        n += 1000000
```
