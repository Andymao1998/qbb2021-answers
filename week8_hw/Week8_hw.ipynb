## generating a binomial distribution assuming a population size of 1000

```python
s = np.random.binomial(2000, 0.4)
s
```

```python
i = 1
P = range(1, 2001)
p = []
s = 0.4
while i <= 2000:
    s = np.random.binomial(2000, s)/2000
    p.append(s)
    i = i+1
```

```python
for i in range(0, 2000):
    if p[i] == 1:
        print(i)
```

```python
plt.scatter(P, p, c = "green", s = 0.5)
plt.xlabel("generation")
plt.ylabel("allele frequency")
plt.title("Initial frequency = 0.4, population = 1000")
plt.show()
```

```python
def generationtester(p, population):
    i = 1
    list = []
    s = p
    while True:
        s = np.random.binomial(2*population, s)/(2*population)
        if s == 0 or s == 1:
            return(i)
            break
        else:
            i = i+1
```

```python
generationtester(0.5, 100)
```

```python
g = []
timer = 1
while timer < 1001:
    g.append(generationtester(0.5, 100))
    timer += 1
```

## For a starting allele frequency of 0.5, vary the population size and produce a plot that shows fixation time vs (N). A reasonable range of population sizes might be 100 to 10 million.

```python
N = []
n = 100
recorder = []
while n <= 10000000:
    recorder.append(generationtester(0.5, n))
    N.append(n)
    print(n)
    if n <= 1000000:
        n += 10000
    if n > 100000:
        n += 100000
```
