## Visualization of the final annotated file

```python
import io
import os
import pandas as pd


def read_vcf(path):
    with open(path, 'r') as f:
        lines = [l for l in f if not l.startswith('##')]
    return pd.read_csv(
        io.StringIO(''.join(lines)),
        dtype={'#CHROM': str, 'POS': int, 'ID': str, 'REF': str, 'ALT': str,
               'QUAL': str, 'FILTER': str, 'INFO': str, 'Sample9': str, 
               'Sample11': str, 'Sample23': str, 'Sample24': str, 'Sample27': str,
               'Sample31': str, 'Sample35': str, 'Sample39': str, 'Sample62': str,
               'Sample63': str},        
        sep='\t'
    ).rename(columns={'#CHROM': 'CHROM'})
```

```python
df = read_vcf("prediction.vcf")
```

## For extracting data from vcf file I choose the samtools plugin bcftools, exact commands are defined in md file. Major call bcftools query -f '%ANN\n' prediction.vcf in terminal

```python
DP = open('read.depth.csv').readlines()
qual = open('qual.csv').readlines()
all_freq = open('allele_freq.csv').readlines()
variant_info = open('variant.info').readlines()
```

### Tidy up quality data

```python
g = []
for i in qual: 
    i = i.replace('\n','')
    g.append(i)
for i in range(0, len(g)):
    g[i] = float(g[i])
```

```python
import statistics as st
st.median(g)
```

### Tidy up read depth data

```python
t = []
for i in DP: 
    i = i.replace('\n','').split(',')
    t.extend(i)
for i in range(0, len(t)):
    t[i] = int(t[i])
```

```python
import statistics as st
st.median(t) 
```

### Tidy up alt allele frequency

```python
all_freq
```

```python
x = []
for i in all_freq: 
    i = i.replace('\n','').split(',')
    x.extend(i)
for i in range(0, len(x)):
    x[i] = float(x[i])
x
```

```python
variant_info[1]
```

### Tidy up variant info, for this section I choose to enumerate all available info for each variant by creating two for loops

```python
v = []
n = []
for i in variant_info: 
    i = i.split(',')
    v.append(i)
for i in range(0, len(v)):
    for j in range(0, len(v[i])):
        n.append(v[i][j].split('|')[0:2])
n
```

```python
side_effect = pd.DataFrame(n, columns = ['gene','info'])
side_effect["info"]
```

```python
side_effect
```

```python
r = side_effect.loc[:,'info'].value_counts()
len(r.index)
```

```python
import matplotlib as ml
from matplotlib import pyplot as plt
fig, ax = plt.subplots(2,2, figsize=(60,130))
ax[0, 0].hist(t, bins =400, color = 'green')
ax[0, 0].set_xlim([0, 180])
ax[0, 0].set_title("Read depth distribution", fontsize=100)
ax[0, 1].hist(g, bins =1000, color = 'yellow')
ax[0, 1].set_xlim([0, 2000])
ax[0, 1].set_ylim([0, 10000])
ax[0, 1].set_title("quality distribution", fontsize=100)
ax[1,0].hist(x, bins =10, color = 'red')
ax[1,0].set_xlim([0, 1])
ax[1,0].set_ylim([0, 15000])
ax[0, 1].set_title("Allele frequency spectrum", fontsize=100)
ax[1 ,1].bar(r.index, height = r )
ax[1, 1].set_title("Summary of the predicted effect", fontsize=100)
#ml.rcParams['font.size']=80
ml.rc('xtick', labelsize=45)
plt.xticks(rotation=90)
ml.rc('ytick', labelsize=100)
fig.savefig('final.pdf')
```
